### Парсинг

```python
dataset = []

# Загружаем описания компаний один раз перед циклом
with open("companies_descriptions.json", "r", encoding="utf-8") as f2:
    companies_descriptions = json.load(f2)

for i, file in enumerate(files):
    with open(os.path.join(data_dir, file), "r", encoding="utf-8") as f:
        data = json.load(f)

    info_data = data["info"]
    base_rate = None
    base_about = None
    base_subs = None
    company_name = os.path.splitext(file)[0]

    if info_data:
        base_rate = info_data.get("rate", float('nan'))
        base_subs = info_data.get("subs", float('nan'))
        base_about = info_data.get("about", float('nan'))
    base_rate = 0 if pd.isna(base_rate) else base_rate
    base_field = companies_descriptions.get(company_name, float('nan'))

    for ref in data["refs"]:
        if not ref:
            continue

        title = ref[0] if ref[0] else float('nan')
        date_info = ref[1]
        stats = ref[2]

        date_string = f'{date_info["day"]} {date_info["month"]} {date_info["time"]}' if date_info else float('nan')
        
        record = {
            "title": title,
            "namecompany": company_name, 
            "description": base_about,
            "rating": base_rate,
            "field": base_field,
            "date": date_string,
            "textpub": title
        }
        dataset.append(record)

df = pd.DataFrame(dataset)
```
