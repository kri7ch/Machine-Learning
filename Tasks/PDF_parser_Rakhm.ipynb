```python
# извлечения текста из PDF-файла
def extract_text_from_pdf(pdf):
    doc = fitz.open(pdf)
    print("Название: ", doc.name)
    text = []
    for current_page in range(len(doc)):
        page = doc.load_page(current_page)
        page_text = page.get_text("text")
        text.append(page_text)
    return text
```

```python
# парсинг текста из PDF
def pars_pdf(text):
    date_publish = "N/A"
    for page in text:
        page_text = page.split('\n')
        page_text = [line for line in page_text if not re.fullmatch(r', ', line)]
        if 'Рейтинг' in page_text:
            num = page_text.index('Рейтинг')
            title = page_text[num+6]
            name = page_text[num+1]
            rating = page_text[num-1]
            description = page_text[num+2]
            category = page_text[num+1]
            for index, item in enumerate(page_text):
                if "назад" in item:
                    date_publish = page_text[index]
                    print(date_publish)
                    if not date_publish.strip()[0].isdigit():
                        date_publish = " ".join(date_publish.split()[1:]).lstrip()
                        print(date_publish)
            content = " ".join(" ".join(text).split("\n")).replace(rating, "").replace(name, "").replace(title, "").replace(description, "").replace(date_publish, "").replace("Рейтинг", "", 1)
            if(date_publish != None):
                break
             
    return  title, name, description, rating, category, date_publish, content
```

```python
# перебор текста из pdf файла, парсинг
Brak = []
Title = []
NameCompany = []
Description = []
Raiting = []
Category = []
DataPublish = []
TextArticle = []
for pdf in all_pdf:
    try:  
        Text = extract_text_from_pdf(pdf)
        t, n, desc, r, cat, date_publish, text = pars_pdf(Text)
        Title.append(t)
        NameCompany.append(n)
        Description.append(desc)
        Raiting.append(r)
        Category.append(cat)
        DataPublish.append(date_publish)
        TextArticle.append(text)
    except Exception as e:
        print(e)
        Brak.append(pdf)
```

### Функции для очистки текста

```python
def remove_rate(text):
    words = text.split()
    if 'Оценка' in words:
        index = words.index('Оценка')
        return ' '.join(words[index + 1:])
    return ' '.join(words)

def remove_average(text):
    match = re.search(r"([А-Яа-яA-Za-z]+\d+)", text)
    if match:
        matched_word = match.group(0)
        words = text.split()
        if matched_word in words:
            i = words.index(matched_word)
            return ' '.join(words[i + 2:])
    return text

def remove_autor(text):
    words = text.split()
    if 'АВТОРОМ' in words:
        for word in ["КАК", "СТАТЬ", "АВТОРОМ"]:
            while word in words:
                words.remove(word)
    for company in ["redmadrobot", "artyomsoft", "TourmalineCore"]:
        if text.startswith(company):
            words = words[len(company) + 1:]
    return ' '.join(words)
```

```python
df['Текст'] = df['Текст'].apply(remove_rate)
df['Текст'] = df['Текст'].apply(remove_average)
df['Текст'] = df['Текст'].apply(remove_autor)
```

* remove_rate — убирает всё до слова "Оценка" включительно
* remove_average — удаляет средние оценки, попадающие в текст
* remove_autor — удаляет фразы, связанные с авторами или специфичными компаниями
